```python
import pandas as pd

# Load your prediction dataset
data = pd.read_csv("C:/Maize_Chainat_C/45Datapoint_Master/5VI_RawData_30Mar_Multiparameter_V8_CV_ALL.csv")

# For demonstration purposes, let's assume your dataset has columns with predictions from the three models: 
# 'xgboost_pred', 'gradient_boosting_pred', and 'catboost_pred'

# Simple Voting
data['simple_voting_pred'] = (data['xgboost_pred'] + data['gradient_boosting_pred'] + data['catboost_pred']) / 3

# Weighted Voting
# Assign weights based on model performance. 
# These weights are placeholders; you should assign them based on your models' performances.
w_xgboost = 0.5  # weight for XGBoost
w_gradient_boosting = 0.3  # weight for Gradient Boosting
w_catboost = 0.2  # weight for CatBoost

data['weighted_voting_pred'] = (w_xgboost * data['xgboost_pred'] +
                                w_gradient_boosting * data['gradient_boosting_pred'] +
                                w_catboost * data['catboost_pred'])

# Save the results
data.to_csv("ensemble_predictions.csv", index=False)

print("Ensemble predictions saved to 'ensemble_predictions.csv'")
```
